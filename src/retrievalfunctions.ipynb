```python
# Content retrieval ordered by how many times a specific word is mentioned in the content of the text. 
# Returns top 3 medical conditions with the most mentions of the searched up word in the content.
import pandas as pd
import re
def search_conditions(query):
    df = pd.read_csv("/Users/timmyg/Downloads/illness.csv")
    df = df.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)
    pd.set_option('display.max_colwidth', 1000) 
    query = query.lower()
    results = df[df["Content"].str.contains(query, case=False, na=False)]
    df_sorted = results[["Title", "Content"]]
    df_sorted["Content"].str.count(query)
    df_sorted_real = df_sorted.sort_values('Content', ascending=False).reset_index(drop=True)
    # lowkey had no clue how to do the snippets part so this is from gpt..if anyone has any mods that would be great!
    def extract_snippet(text):
        matches = re.findall(fr'(?i).{{0,200}}\b{query}\b.{{0,200}}', text)
        if matches:
            return " ... ".join(m.strip() for m in matches)
        else:
            return ""
    df_sorted_real["Search Relevant Content"] = df_sorted_real["Content"].apply(extract_snippet)
    return df_sorted_real.head(3)

search_conditions("headache")
```

```python
# Alternatively, if the title contains the content searched up it's returned in this function.
import pandas as pd
def search_conditions(query):
    df = pd.read_csv("/Users/timmyg/Downloads/illness.csv")
    df = df.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)
    pd.set_option('display.max_colwidth', 1000) 
    query = query.lower()
    results = df[df["Title"].str.contains(query, case=False, na=False)]
    return results[["Title", "Content"]]

search_conditions("Flu")
```
